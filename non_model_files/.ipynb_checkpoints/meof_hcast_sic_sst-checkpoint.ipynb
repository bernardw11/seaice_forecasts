## Using the standard eofs package for the MEOF analysis (converting to numpy array and back)

## hmmmmmmmmmmmmmmmmmmmmmmmmmm now do we want eofsAsCovariance here or just eofs. that is the question to ask cuihua and xiaojun.

```python
# Convert dataarrays into numpy arrays to do the MEOF analysis
np_sic_anom = sic_anom.to_numpy()
np_sst_anom = sst_anom.to_numpy()
solver = MultivariateEof([np_sic_anom, np_sst_anom],
                         weights = [sic_wgts, sst_wgts])

#___This is the thing we change here.
modes = 3

np_eofs_list = solver.eofs(neofs=modes)
#np_eofs_list = solver.eofsAsCovariance(neofs=modes) # datasets: 2, mode: mode, y: 20, x: 180
# np_eofs_list: contains a list of EOFs for every variable. 
#   2, mode, y, x.
#   first one: 0, mode, 80, 360
#   second one: 1, mode, 20, 180
np_pcs = solver.pcs(npcs=modes) # tdim: 518, mode: mode
np_eigs = solver.eigenvalues(neigs=modes) 
sum_eigs = solver.eigenvalues().sum()
percent_var = np_eigs.sum() / solver.eigenvalues().sum()
```

```python
# Convert np arrays back into dataarrays
sic_lat_bins = np.arange(-89.5, -49.5, 0.5) #Y: 80
sic_lon_bins = np.arange(0., 360, 1) # X: 360
sst_lat_bins = np.arange(-88., -49, 2)
sst_lon_bins = np.arange(0., 359, 2)

sic_eofs = xr.DataArray(np_eofs_list[0],
                       coords = [np.arange(0, modes), sic_lat_bins, sic_lon_bins],
                        dims=['mode', 'y', 'x'])
sst_eofs = xr.DataArray(np_eofs_list[1],
                       coords = [np.arange(0, modes), sst_lat_bins, sst_lon_bins],
                        dims=['mode', 'y', 'x'])
pcs = xr.DataArray(np_pcs, coords=[sic_anom.tdim, np.arange(0, modes)],
                  dims=['tdim', 'mode'])
```

```python
# Plot EOFs 
fig = plt.figure(figsize=(20,10))
for mode in range(modes):
    #SIC EOFs
    ax = plt.subplot(2, 5, mode + 1, projection=sp)
    im = sic_eofs[mode].plot(transform=pc, add_colorbar=False)
    ax.coastlines()
    ax.gridlines(draw_labels=True)
    ax.set_extent([-3950000., 3950000., 4350000., -3950000.], crs=sp)
    ax.set_title('mode = ' + str(mode) + ', %var: ' + str(round(np_eigs[mode]/sum_eigs * 100, 2)))
    
    #SST EOFs
    ax = plt.subplot(2, 5, mode + 6, projection=sp)
    im = sst_eofs[mode].plot(transform=pc, add_colorbar=False)
    ax.coastlines()
    ax.gridlines(draw_labels=True)
    ax.set_extent([-3950000., 3950000., 4350000., -3950000.], crs=sp)
    
plt.suptitle('LEADING EOFS')
# eofs.plot(x='x',y='y',col='mode',col_wrap=modes,add_colorbar=0)

# Plot corresponding PCs
pcs.plot(x='tdim',col='mode',col_wrap=modes);
print(solver.eigenvalues(neigs=modes))
```

### GETTING THE TRANSITION MATRIX steps:
1. Build 12 P_i's: nc x #yrs x 12mon matrix. each month layer is P_i: nc x #yrs
2. Compute d = <P_i P_i^t>
3. Compute c = <P_i+1 P_i^t>
4. Invert d to get d_i = <P_i P_i^t>^-1
5. Matrix multiply c and d_i together to get a, for that month out of 12 months.
6. Do it for all 12 months and built out the 12 x nc x nc matrix of transition matrices

```python
pcs_monthly = pcs.groupby('tdim.month')
# indices 1-12 of pcs_monthly correspond to the pc of the month
yrs = int(len(pcs.tdim)/12 - 1)
# We have 43 years, but we want to loop thru 42 b/c we need the last December to correspond to a January of the last year.
```

```python
# Declare aa
aa_py = np.empty([12, modes, modes])

for month in range(1, 13):
    #declare d, c as nc x nc matrices.
    d = np.empty([modes, modes])
    c = np.empty([modes, modes])
    
    # select the PC's of that specific month, a yearly time series.
    p_i = pcs_monthly[month].drop_isel(tdim=42)
    # Looping through each combination of modes (ex. 5x5):
    for i in range(modes):
        for j in range(modes):    
            # Build d: dot product the PCs of the two modes. Get the mean.
            d[i, j] = np.dot(p_i.sel(mode=i), p_i.sel(mode=j)) / yrs
            
            # Build c: <P_i+1 * P_i>
            # If it is December: dot product the PCts of the decembers w/ the PCts of the next jan's
            if month == 12:
                p_i2 =  pcs_monthly[1].drop_isel(tdim=0)
                c[i, j] = np.dot(p_i2.sel(mode=i), p_i.sel(mode=j)) / yrs
            # If it is not December:
            else:
                p_i2 = pcs_monthly[month + 1].drop_isel(tdim=42)
                c[i, j] = np.dot(p_i2.sel(mode=i), p_i.sel(mode=j)) / yrs
            
    # invert d to get d_i
    d_i = np.linalg.inv(d)
    
    # using d_i and c, calculate a (transition matrix for the month.)
    a = np.matmul(c, d_i)
    
    # add to the big matrix of transition matrices.
    aa_py[month - 1] = a
```
