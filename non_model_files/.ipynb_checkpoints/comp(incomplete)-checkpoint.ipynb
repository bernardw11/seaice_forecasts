```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import datetime
from dateutil.relativedelta import relativedelta

from cartopy import crs as ccrs, feature as cfeature
import warnings
warnings.filterwarnings('ignore')

import xarray as xr
import glob, os
```

```python
# Data: 1979-2021 for now.
```

```python
# Taking the SST dataset: up to -50 degrees south.
# Small grid: 180x15 grid

filename = '/d6/bxw2101/ersst_monthly/all_sst.nc'
sst_ds = xr.open_dataset(filename, decode_times=False)
filename = '/d6/bxw2101/ersst_monthly/all_sst_anom.nc'
anomsst_ds = xr.open_dataset(filename, decode_times=False)

date_1 = datetime.datetime.strptime("01/01/1960", "%m/%d/%Y")
def convert_times(x):
  return date_1 + relativedelta(months=int(x))
convert_times_v = np.vectorize(convert_times)

sst_ds = sst_ds.assign_coords(T=convert_times_v(sst_ds.T))
sst_ds = sst_ds.rename({'T': 'tdim'})
anomsst_ds = anomsst_ds.assign_coords(T=convert_times_v(anomsst_ds.T))
anomsst_ds = anomsst_ds.rename({'T': 'tdim'})

sst_ds_crop = sst_ds.sel(Y=slice(-88, -50))
sst_ds_crop = sst_ds_crop.sel(tdim=slice('1979-01-01', '2021-12-01'))
sst_var = sst_ds_crop.sst
anomsst_ds_crop = anomsst_ds.sel(Y=slice(-88, -50))
anomsst_ds_crop = anomsst_ds_crop.sel(tdim=slice('1979-01-01', '2021-12-01'))
anom_var = anomsst_ds_crop.anom
```

```python
# Taking the SIC dataset: 
filename = '/d6/bxw2101/seaice_conc_monthly/all_seaice_conc_monthly.nc'
ds = xr.open_dataset(filename)
ds = ds.rename(time='tdim')
ds = ds.rename(xgrid='x')
ds = ds.rename(ygrid='y')
filename = '/d6/bxw2101/seaice_conc_monthly/grid_files/NSIDC0771_LatLon_PS_S25km_v1.0.nc'
psgds = xr.open_dataset(filename)
ds = ds.assign_coords(nav_lat=psgds.latitude)
sic_ds = ds.assign_coords(nav_lon=psgds.longitude)
sic_var_b4 = sic_ds.cdr_seaice_conc_monthly
sic_var = sic_var_b4.isel(x=slice(0,316), y=slice(0,332))
sic_var = sic_var.where((sic_var < 2.51)) # Must mask some of the data because the flag values are 251 to 255.
sic_mon = sic_var.groupby('tdim.month')
sic_clim = sic_mon.mean(dim='tdim')
sic_anom = sic_mon - sic_clim
```

```python
# Taking the OHC dataset:
filename = '/d6/bxw2101/oras5_ohc/all_ohc_monthly.nc'
ds = xr.open_dataset(filename, decode_times=False)
date_1 = datetime.datetime.strptime("01/16/1978", "%m/%d/%Y")
def convert_times(x):
  return (date_1 + relativedelta(hours=int(x))).replace(day=1)
convert_times_v = np.vectorize(convert_times)
ds = ds.assign_coords(time_counter=convert_times_v(ds.time_counter))
ohc_ds = ds.rename({'time_counter': 'tdim'})
ohc = ohc_ds.sohtc300
clim_period = ohc_ds.sel(tdim=slice('1979-01-01', '2021-12-01'))
clim_ohc = clim_period.sohtc300
ohc_mon = ohc.groupby('tdim.month')
ohc_clim = clim_ohc.groupby('tdim.month').mean(dim='tdim')
ohc_anom = ohc_mon - ohc_clim
```

```python
mean_sic_anom = sic_anom.mean(dim=['x','y'])
sic_rolling = mean_sic_anom.rolling(tdim=5, center=True).mean()
nsic_rolling = sic_rolling / mean_sic_anom.std()

datasetanom = anom_var.mean(dim=['X', 'Y'])
sst_rolling = datasetanom.rolling(tdim=5, center=True).mean()
nsst_rolling = sst_rolling / datasetanom.std()

mean_ohc_anom = ohc_anom.mean(dim=['x', 'y'])
ohc_rolling = mean_ohc_anom.rolling(tdim=5, center=True).mean()
nohc_rolling = ohc_rolling / mean_ohc_anom.std()
```

```python
nsic_rolling.plot(size=7)
nsst_rolling.plot()
plt.legend([ 'Mean SIC Anomaly', 'Mean SST Anomaly'])
plt.title('Anomaly time series of SIC, SST');
plt.ylabel('Anomaly, divided by its standard deviation')
```

```python
nsic_rolling.plot(size=7)
nohc_rolling.plot()
plt.legend([ 'Mean SIC Anomaly', 'Mean OHC Anomaly'])
plt.title('Anomaly time series of SIC, OHC');
plt.ylabel('Anomaly, divided by its standard deviation')


#Get sea ice thickness dataset!
```
