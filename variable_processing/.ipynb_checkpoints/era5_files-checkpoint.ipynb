# This notebook turns all four era5 files (sat, sp, 850hPau, 850hPav) and converts the lat/lon 0.25x0.25 grid to lat/lon 2x2 grid.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1 import ImageGrid
import datetime
from dateutil.relativedelta import relativedelta

from cartopy import crs as ccrs, feature as cfeature
import warnings
warnings.filterwarnings('ignore')

import netCDF4
from netCDF4 import Dataset

import xarray as xr
import glob, os
#print(xr.__version__)

sp = ccrs.SouthPolarStereo()
pc = ccrs.PlateCarree() # The OG Data is in pc projection.
```

```python
filename = '/d6/bxw2101/era5/sat.nc'
sat_ds = xr.open_dataset(filename)
filename = '/d6/bxw2101/era5/gp.nc'
gp_ds = xr.open_dataset(filename)
filename = '/d6/bxw2101/era5/uwind.nc'
uwind_ds = xr.open_dataset(filename)
filename = '/d6/bxw2101/era5/vwind.nc'
vwind_ds = xr.open_dataset(filename)
filename = '/d6/bxw2101/era5/sensible.nc'
sensible_ds = xr.open_dataset(filename)
filename = '/d6/bxw2101/era5/latent.nc'
latent_ds = xr.open_dataset(filename)
```

```python
latent_ds
```

```python
lat_bins = np.arange(-88., -49, 2)
tmp_lon_bins = np.arange(-180., 179, 2)
lon_bins_ordered = np.arange(0., 359, 2)
lon_bins = np.append(np.arange(180., 359, 2), np.arange(0., 179, 2))
```

```python
# Crop the 0.25x0.25 grids to 2.0x2.0 grids, put the longitudes into 0.0-358.0, rename the coords&dims, fillna.
# (takes a min)
sat_crop_ds = sat_ds.sel(latitude=lat_bins).sel(longitude=tmp_lon_bins).assign_coords(longitude=lon_bins).sortby("longitude").rename({'longitude': 'x', 'latitude': 'y', 'time': 'tdim'}).fillna(0)
gp_crop_ds = gp_ds.sel(latitude=lat_bins).sel(longitude=tmp_lon_bins).assign_coords(longitude=lon_bins).sortby("longitude").rename({'longitude': 'x', 'latitude': 'y', 'time': 'tdim'}).fillna(0)
uwind_crop_ds = uwind_ds.sel(latitude=lat_bins).sel(longitude=tmp_lon_bins).assign_coords(longitude=lon_bins).sortby("longitude").rename({'longitude': 'x', 'latitude': 'y', 'time': 'tdim'}).fillna(0)
vwind_crop_ds = vwind_ds.sel(latitude=lat_bins).sel(longitude=tmp_lon_bins).assign_coords(longitude=lon_bins).sortby("longitude").rename({'longitude': 'x', 'latitude': 'y', 'time': 'tdim'}).fillna(0)
sensible_crop_ds = sensible_ds.sel(latitude=lat_bins).sel(longitude=tmp_lon_bins).assign_coords(longitude=lon_bins).sortby("longitude").rename({'longitude': 'x', 'latitude': 'y', 'time': 'tdim'}).fillna(0)
latent_crop_ds = latent_ds.sel(latitude=lat_bins).sel(longitude=tmp_lon_bins).assign_coords(longitude=lon_bins).sortby("longitude").rename({'longitude': 'x', 'latitude': 'y', 'time': 'tdim'}).fillna(0)
```

```python
# Get the climatological anomalies for all of these datasets.
sat = sat_crop_ds.t2m-273.15
sat_anom = sat.groupby('tdim.month') - sat.groupby('tdim.month').mean(dim='tdim')

gph_850 = gp_crop_ds.z/9.80665
gph_850_anom = gph_850.groupby('tdim.month') - gph_850.groupby('tdim.month').mean(dim='tdim')

uwind_850 = uwind_crop_ds.u
uwind_850_anom = uwind_850.groupby('tdim.month') - uwind_850.groupby('tdim.month').mean(dim='tdim')

vwind_850 = vwind_crop_ds.v
vwind_850_anom = vwind_850.groupby('tdim.month') - vwind_850.groupby('tdim.month').mean(dim='tdim')

sthf = sensible_crop_ds.sshf + latent_crop_ds.slhf
sthf_anom = sthf.groupby('tdim.month') - sthf.groupby('tdim.month').mean(dim='tdim')
```

### Save the combined era5 dataset to a file.

```python
era5_vars = xr.merge([sat_anom, gph_850_anom, uwind_850_anom, vwind_850_anom]).drop_vars('month')
era5_vars.to_netcdf('/d6/bxw2101/combined_netcdf_files/era5_anom_monthly_2x2.nc', mode='w',format='NETCDF4')
```
